Traiter les valeurs aberrantes de SupportTickets (F35)

```python
import numpy as np
```

```python
print("\n--- TRAITEMENT DES VALEURS ABERRANTES ---")
print("Avant traitement - SupportTickets (F35):")
print(observations['F35'].describe())

# Remplacer -1 par NaN (valeur manquante)
observations.loc[observations['F35'] == -1, 'F35'] = np.nan
# Remplacer 999 par la médiane ou une valeur max raisonnable
median_tickets = observations[observations['F35'] < 100]['F35'].median()
observations.loc[observations['F35'] > 50, 'F35'] = median_tickets

print("\nAprès traitement - SupportTickets (F35):")
print(observations['F35'].describe())
```

 Traiter SatisfactionScore (F36)

```python
print("\n--- TRAITEMENT DE SATISFACTION SCORE ---")
print("Avant traitement - Satisfaction (F36):")
print(observations['F36'].value_counts().sort_index())

# Remplacer -1 par NaN
observations.loc[observations['F36'] == -1, 'F36'] = np.nan
# Remplacer 99 par la médiane
median_satisfaction = observations[observations['F36'] <= 5]['F36'].median()
observations.loc[observations['F36'] > 5, 'F36'] = median_satisfaction

print("\nAprès traitement - Satisfaction (F36):")
print(observations['F36'].describe())
```

Traiter les valeurs manquantes de l'âge (F31)

```python
print("\n--- TRAITEMENT DE L'ÂGE ---")
print(f"Valeurs manquantes avant : {observations['F31'].isnull().sum()}")

# Imputation par la médiane (robuste aux outliers)
age_median = observations['F31'].median()
observations['F31'].fillna(age_median, inplace=True)

print(f"Valeurs manquantes après : {observations['F31'].isnull().sum()}")
print(f"Médiane utilisée : {age_median}")
```

Encodage des variables catégorielles

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
```

```python

print("\n--- ENCODAGE DES VARIABLES CATÉGORIELLES ---")
print("Variables catégorielles à encoder :")
print(categorical_cols.tolist())

# Créer une copie pour éviter les warnings
observations_encoded = observations.copy()

# Label Encoding pour les variables ordinales
ordinal_cols = ['F37', 'F38', 'F40', 'F43', 'F44', 'F46']  # À ajuster selon votre analyse
for col in ordinal_cols:
    if col in categorical_cols:
        le = LabelEncoder()
        observations_encoded[col] = le.fit_transform(observations_encoded[col].astype(str))
        print(f"Encodé: {col}")

# One-Hot Encoding pour les variables nominales
nominal_cols = [col for col in categorical_cols if col not in ordinal_cols]
observations_encoded = pd.get_dummies(observations_encoded, columns=nominal_cols, drop_first=True)

print(f"\nNouvelle dimension après encodage : {observations_encoded.shape}")
```

Normalisation des features numériques

```python
from sklearn.preprocessing import StandardScaler

# Identifier les colonnes numériques (maintenant inclut les encodées)
feature_cols = [col for col in observations_encoded.columns if col != 'Churn']

# Standardisation
scaler = StandardScaler()
observations_encoded[feature_cols] = scaler.fit_transform(observations_encoded[feature_cols])

print("\n--- APERÇU APRÈS NORMALISATION ---")
print(observations_encoded[feature_cols].describe().round(2))
```

Préparation finale pour la modélisation

```python
# 7. Préparation finale pour la modélisation
from sklearn.model_selection import train_test_split

# Séparer X et y
X = observations_encoded.drop('Churn', axis=1)
y = observations_encoded['Churn']

# Split train/test (80/20) avec stratification
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.20, random_state=42, stratify=y
)

print(f"\n--- SPLIT TRAIN/TEST ---")
print(f"X_train shape: {X_train.shape}")
print(f"X_test shape: {X_test.shape}")
print(f"y_train distribution:\n{y_train.value_counts(normalize=True)}")
print(f"y_test distribution:\n{y_test.value_counts(normalize=True)}")
```

Sauvegarde dans data/train_test/

```python
X_train.to_csv('../data/train_test/X_train.csv', index=False)
X_test.to_csv('../data/train_test/X_test.csv', index=False)
y_train.to_csv('../data/train_test/y_train.csv', index=False)
y_test.to_csv('../data/train_test/y_test.csv', index=False)
print("\n Données sauvegardées dans data/train_test/")
```
